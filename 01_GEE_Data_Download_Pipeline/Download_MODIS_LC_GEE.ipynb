```python

import ee
import geemap

# Initialize Earth Engine

ee.Authenticate()
ee.Initialize()
# Define region of interest
region = ee.Geometry.Polygon([
    [
        [-124.3998113967669, 46.328967879608506],
        [-124.3998113967669, 41.620634777502005],
        [-116.4457098342669, 41.620634777502005],
        [-116.4457098342669, 46.328967879608506],
        [-124.3998113967669, 46.328967879608506]
    ]
])

```

```python
# Load MODIS Land Cover (MCD12Q1) 2020 and select LC_Type1
modis_lc = ee.ImageCollection("MODIS/006/MCD12Q1") \
    .filterDate('2020-01-01', '2020-12-31') \
    .first() \
    .select('LC_Type1') \
    .clip(region)

# Define land cover visualization parameters
landcover_vis = {
    'min': 0,
    'max': 17,
    'palette': [
        '05450a', '086a10', '54a708', '78d203', '009900', 'c6b044',
        'dcd159', 'dade48', 'fbff13', 'b6ff05', '27ff87', 'c24f44',
        'a5a5a5', 'ff6d4c', '69fff8', 'f9ffa4', '1c0dff', 'ffffff'
    ]
}

# Create map and center on region
Map = geemap.Map(center=[44, -120], zoom=6)
Map.addLayer(modis_lc, landcover_vis, 'MODIS Land Cover 2020')
Map.addLayer(region, {}, 'ROI')
Map

# Export to Drive
task = ee.batch.Export.image.toDrive(
    image=modis_lc,
    description='MODIS_LandCover_2020_PNW',
    folder='GEE_Exports',
    fileNamePrefix='modis_landcover_2020_pnw',
    region=region,
    scale=500,
    #crs='EPSG:4326',
    maxPixels=1e13
)
task.start()
print("✅ Export started. Check Google Drive > GEE_Exports")
```
